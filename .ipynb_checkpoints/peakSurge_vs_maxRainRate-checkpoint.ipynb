```python
import sys

sys.path.append(r'C:\Users\lelise\Documents\GitHub\flood_model_carolinas\synthetic_tc_cmpdfld')
import matplotlib as mpl
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib import colors, patheffects
import hydromt
from syntheticTC_utils import *
```

```python
data_catalog_ymls = [r'Z:\Data-Expansion\users\lelise\data\data_catalog_SFINCS_Carolinas.yml', r'Z:\Data-Expansion\users\lelise\data\data_catalog_BASE_Carolinas.yml']
cat = hydromt.DataCatalog(data_libs=data_catalog_ymls)

# Load geographic data for plotting
mod_domain = cat.get_geodataframe(data_like='enc_domain_HUC6_clipped').to_crs(4326)
mod_domain.set_index('Name', inplace=True)
rivers = cat.get_geodataframe(data_like='carolinas_nhd_area_rivers').to_crs(4326).clip(mod_domain)
adcirc_gages = cat.get_geodataframe(data_like=r'Z:\Data-Expansion\users\lelise\Chapter3\NCEP_Reanalysis\stormTide\coastline_lores_NCSC.shp').to_crs(4326)
adcirc_gages.set_index('gage_id', inplace=True)

# Set output directory
outdir = r'Z:\Data-Expansion\users\lelise\Chapter3\NCEP_Reanalysis\stormHazards'
```

```python
# Map the basins and gage points
fig, ax = plt.subplots(
    nrows=1, ncols=1,
    figsize=(6, 4.5),
    tight_layout=True)

adcirc_gages.plot(ax=ax, color='red')
rivers.plot(ax=ax, color='blue')
mod_domain.plot(ax=ax, color='none', edgecolor='black')

# Add title and save figure
ax.set_title('')
ax.yaxis.set_visible(True)
ax.xaxis.set_visible(True)
ax.ticklabel_format(style='sci', useOffset=False)
ax.set_aspect('equal')

# Label gages
for label, grow in adcirc_gages.iterrows():
    ann_kwargs = dict(xytext=(-25, 0), textcoords="offset points", zorder=4,
                      path_effects=[
                          patheffects.Stroke(linewidth=1.5, foreground="white", alpha=0.5),
                          patheffects.Normal(), ], )
    x, y = grow.geometry.centroid.x, grow.geometry.centroid.y
    ax.annotate(f'{label}', xy=(x, y), weight='normal', **ann_kwargs)

for label, grow in mod_domain.iterrows():
    ann_kwargs = dict(xytext=(-15, 15), textcoords="offset points", zorder=4,
                      path_effects=[
                          patheffects.Stroke(linewidth=1.5, foreground="white", alpha=1),
                          patheffects.Normal(), ], )
    x, y = grow.geometry.centroid.x, grow.geometry.centroid.y
    ax.annotate(f'{label}', xy=(x, y), weight='normal', **ann_kwargs)

plt.savefig(os.path.join(outdir, 'basin_gage_map.jpg'), dpi=255)
```

```python
os.chdir(r'Z:\Data-Expansion\users\lelise\Chapter3\NCEP_Reanalysis')

# Load the ADCIRC gage peaks
gage_peaks = pd.read_csv('.\stormTide\gage_peaks.csv', index_col=0)
print(gage_peaks.columns)

# Get a list of the TCs we are looking at
selected_tcs = gage_peaks.index.tolist()

# Load the precip stats calculated for each TC by basin
maxRainRate = pd.read_csv(r'.\rain\TC_maxRainRate_by_basin.csv', index_col=0)
print(maxRainRate.columns)
avgTotP = pd.read_csv(r'.\rain\basinMaxTotPrecip_by_basin.csv', index_col=0)

# Load the wind stats calculated for each TC by basin
maxVmax = pd.read_csv(r'.\wind\TC_Vmax_by_basin.csv', index_col=0)
```

```python
gages = [188, 195, 203, 206, 199]
basins = ['Lower Pee Dee','Cape Fear','Neuse','Pamlico','Onslow Bay']
dfs_plot = []
for i in range(len(basins)):
    gage = gages[i]
    basin = basins[i]

    # Get the surge peaks at the selected ADCIRC gage
    stormTide_gage_peaks = gage_peaks[str(gage)]
    
    # Get the basin max rain rate
    df = maxRainRate[basin].copy()
    d1 = maxVmax[basin].copy()
    
    # Combine into a single dataframe for plotting
    df2 = pd.concat(objs=[df, d1, stormTide_gage_peaks], axis=1, ignore_index=False)
    df2.dropna(how='any',axis=0, inplace=True)
    df2.columns = ['Max Rain Rate', 'Vmax', 'Max Surge']
    dfs_plot.append(df2)
```

```python
# Setup Plot Info
font = {'family': 'Arial', 'size': 10}
mpl.rc('font', **font)
mpl.rcParams.update({'axes.titlesize': 10})

nrow, ncol = 3, 2
n_subplots = nrow * ncol
first_in_row = np.arange(0, n_subplots, ncol)
last_row = np.arange(n_subplots - ncol, n_subplots, 1)
fig, axs = plt.subplots(nrows=nrow,
                        ncols=ncol,
                        figsize=(5.5, 6),
                        constrained_layout=True,
                        sharey=True, 
                        sharex=False
                       )
axs = axs.flatten()
axs[-1].set_visible(False)

for i in range(len(dfs_plot)):
    ax = axs[i]
    df_plot = dfs_plot[i]
    basin = basins[i]

    df_plot.plot.scatter(x='Max Surge',
                         y='Max Rain Rate',
                         c ='Vmax',
                         ax=ax, 
                         marker='.',
                         colormap='viridis'
                        )
    ax.set_xlim([-0.1,4.1])
    ax.set_ylabel('Max Rain\n Rate (mm/hr)')
    ax.set_xlabel(f'Max Surge (m) at Pt {gages[i]}')
    ax.set_title(f'{basin} (n={len(df_plot)})')
plt.subplots_adjust(wspace=0, hspace=0)
plt.margins(x=0, y=0)
plt.savefig(os.path.join(outdir, 'basin_gage_maxRR_vs_maxSurge_vs_Vmax.jpg'), dpi=255)
```
