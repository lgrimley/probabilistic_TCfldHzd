```python
import os
import geopandas as gpd
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import matplotlib as mpl
from hydromt_sfincs import SfincsModel
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib import colors, patheffects
```

```python
os.chdir(r'Z:\Data-Expansion\users\lelise\Chapter3\NCEP_Reanalysis\stormTide')
fname = 'coastline_lores'  # ADCIRC water level locations (low resolution)

# Load in the ADCIRC water level output locations and create a geodataframe
wl_gages = sio.loadmat(f'{fname}.mat')
xcoords = np.concatenate(wl_gages['xvec'], axis=0)
ycoords= np.concatenate(wl_gages['yvec'], axis=0)
gage_id = np.arange(start=1, stop=len(xcoords)+1, step=1)
gages = gpd.GeoDataFrame(geometry=gpd.points_from_xy(x=xcoords, y=ycoords, crs=4326))
gages['gage_id'] = gage_id
gages.index = gage_id
```

```python
# Load in the SFINCS model
model_root = r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter1_FlorenceValidation\sfincs_models\mod_v4_flor\ENC_200m_sbg5m_avgN_adv1_eff75'
mod = SfincsModel(model_root, mode='r')
```

```python
# Select the gages that are within a buffer of the area of interest (aoi)
domain_buffer_km = 15 # km
gages = gages.to_crs(mod.region.crs)  # needs to be in the same CRS as the aoi
aoi_buff = mod.region.buffer(distance=domain_buffer_km*1000)  # create a buffer around the aoi
select_gages = gages.clip(mask = aoi_buff)  # clip the gage data using the buffer
print(f'Points selected: {len(select_gages)} out of {len(gages)}')
print(select_gages)
```

```python
# Setup some things for plotting
cmap = mpl.cm.binary  # colormap black/white
bounds = [0, 15, 30, 45, 60, 120, 200]  # elevation breaks
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='both')
wkt = mod.grid['dep'].raster.crs.to_wkt()
utm_zone = mod.grid['dep'].raster.crs.to_wkt().split("UTM zone ")[1][:3]
utm = ccrs.UTM(int(utm_zone[:2]), "S" in utm_zone)
extent = np.array(mod.region.buffer(domain_buffer_km*1000*2).total_bounds)[[0, 2, 1, 3]]
```

```python
fig, ax = plt.subplots(
    nrows=1, ncols=1,
    figsize=(6.2, 4.5),
    subplot_kw={'projection': utm},
    tight_layout=True)

dem = mod.grid['dep'].plot(ax=ax, cmap=cmap, norm=norm, add_colorbar=False, zorder=1)
gages.to_crs(mod.region.crs).plot(ax=ax, color='red')
select_gages.plot(ax=ax, color='blue')
aoi_buff.plot(ax=ax, color='none', edgecolor='maroon')

# Update extent of figure
minx, miny, maxx, maxy = extent
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)

# Add DEM colorbar
cbar = fig.colorbar(dem, ax=ax, shrink=0.5,
                    extend='both', spacing='uniform',
                    label='Elevation\n(m+NAVD88)',
                    anchor=(0.1, 0.6))

# Add title and save figure
ax.set_extent(extent, crs=utm)
ax.set_title('')
ax.set_ylabel(f"Y Coord UTM zone {utm_zone} (m)")
ax.yaxis.set_visible(True)
ax.set_xlabel(f"X Coord UTM zone {utm_zone} (m)")
ax.xaxis.set_visible(True)
ax.ticklabel_format(style='sci', useOffset=False)
ax.set_aspect('equal')

# Label gages
for label, grow in select_gages.iterrows():
    ann_kwargs = dict(xytext=(-25, 0), textcoords="offset points", zorder=4,
                      path_effects=[
                          patheffects.Stroke(linewidth=1.5, foreground="white", alpha=1),
                          patheffects.Normal(), ], )
    x, y = grow.geometry.centroid.x, grow.geometry.centroid.y
    ax.annotate(f'{label}', xy=(x, y), weight='normal', **ann_kwargs)

plt.subplots_adjust(wspace=0, hspace=0)
plt.margins(x=0, y=0)
plt.savefig('ADCIRC gage locations for boundary conditions.png')
```

```python
s = select_gages.to_crs(4326)
s.to_file(r'coastline_lores_NCSC.shp')

s['x'] = s.geometry.x.values
s['y'] = s.geometry.y.values
s.drop(columns='geometry', inplace=True)
s.to_csv(r'coastline_lores_NCSC.csv', index=False)
```
