```python
hmin = 0.05
tc_ids = [x for x in attr_ds.tc_id.values.tolist() if 'SLR' not in x]
fld_area = pd.DataFrame()
for tc_id in tc_ids:
    # Get storm track
    tc_track_gdf = tc_tracks_polyline[tc_tracks_polyline['tc_id'] == int(tc_id)]
    tc_track_gdf = tc_track_gdf.to_crs(32617)
    
    ####### Sea level rise
    attr = attr_ds.sel(tc_id=f'{str(tc_id)}_SLR112cm')
    da_diff_slr = attr['zsmax_diff']
    compound_mask_slr = da_diff_slr > hmin
    zsmax_ds_slr = zsmax_ds.sel(tc_id=f'{str(tc_id)}_SLR112cm', scenario='compound')['zsmax']
    zsmax_ds_slr = xr.where(zsmax_ds_slr > hmin, 1, np.nan)

    ####### No Sea level rise
    attr = attr_ds.sel(tc_id=str(tc_id))
    da_diff_noslr = attr['zsmax_diff']
    compound_mask_noslr = da_diff_noslr > hmin
    
    # diff == 3 is compound in SLR only, diff == 2 is compound in both, diff == 1 is compound in no SLR only
    cmpd_extent_slr = xr.where(compound_mask_slr, x=3, y=0)
    cmpd_extent_noslr = xr.where(compound_mask_noslr, x=1, y=0)
    diff_cat = (cmpd_extent_slr - cmpd_extent_noslr).compute()
    
    # SLR, No SLR, Both raster
    diff_cat = xr.where(diff_cat == -1, 1, diff_cat) # Compound w/out SLR only
    diff_cat = diff_cat.where(diff_cat > 0)
    
    unique_codes, cell_counts = np.unique(diff_cat.data, return_counts=True)
    fld_area1 = cell_counts.copy()
    res = 200  # grid cell resolution in meters
    fld_area1 = fld_area1 * (res * res) / (1000 ** 2)  # square km
    
    # Cleanup the dataframe
    fld_area1 = pd.DataFrame(fld_area1).T
    fld_area1.columns = unique_codes
    expected = pd.DataFrame(columns=[1.0, 2.0, 3.0, np.NAN])
    fld_area1 = pd.concat(objs=[expected, fld_area1], axis=0, ignore_index=True)
    fld_area1.columns = ['NoSLR', 'Both', 'SLR', 'Inactive']
    fld_area = pd.concat(objs=[fld_area, fld_area1], axis=0, ignore_index=True)

    plot_please = True
    if plot_please is True:
        ####### PLOT
        tick_labels = ['w/out SLR', 'Both', 'w/ SLR']
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 4), subplot_kw={'projection': utm}, tight_layout=True)
        
        cmap = mpl.colors.ListedColormap(['lightgrey'])
        bounds = [0,1]
        norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='neither')
        cs1 = zsmax_ds_slr.plot(ax=ax, cmap=cmap, norm=norm, add_colorbar=False, zorder=2, alpha=0.5)

        cmap = mpl.colors.ListedColormap(['magenta', 'royalblue', 'darkorange'])
        bounds = [1, 2, 3, 4]
        norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='neither')
        cs = diff_cat.plot(ax=ax, cmap=cmap, norm=norm, add_colorbar=False, zorder=2, alpha=1)
        
        major_rivers_clip.plot(ax=ax, color='none', edgecolor='slategrey', linewidth=0.25, zorder=0, alpha=0.75)
        nc_major_rivers_clip.plot(ax=ax, color='none', edgecolor='slategrey', linewidth=0.25, zorder=0, alpha=0.75)
        coastal_wb_clip.plot(ax=ax, color='none', edgecolor='slategrey', linewidth=0.25, zorder=0, alpha=0.75)
        mod.region.plot(ax=ax, color='none', edgecolor='black', linewidth=0.75, zorder=3, alpha=1)
        tc_track_gdf.geometry.plot(ax=ax, color='black', linewidth=2, zorder=3)
    
        minx, miny, maxx, maxy = mod.region.total_bounds
        ax.set_xlim(minx, maxx)
        ax.set_ylim(miny, maxy)
        
        ax.set_title(f'Compound Extent TC{tc_id} (hmin:{hmin}m)')
        pos0 = ax.get_position()  # get the original position
        cax = fig.add_axes([pos0.x1 + 0.02, pos0.y0 + pos0.height * 0.1, 0.05, pos0.y0 + pos0.height * 0.8])
        cbar = fig.colorbar(cs, cax=cax, orientation='vertical', extend='neither', ticks=[1.5, 2.5, 3.5])
        cbar.ax.set_yticklabels(labels=tick_labels)
        
        plt.subplots_adjust(wspace=0, hspace=0)
        plt.margins(x=0, y=0)
        plt.savefig(rf'.\SLR_analysis\max_cmpd_extent_{tc_id}_hmin{hmin}.jpg', bbox_inches='tight', dpi=300)
        plt.close()
    
    print(tc_id)
```
